```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../datasets/Titanic-Dataset.csv')
```

```python
plt.figure(figsize=(8,5))
ax=sns.barplot(x='Sex', y='Survived', data=df, hue='Sex',palette={'male':'#4C72B0', 'female':'#55A868'}, errorbar=None)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height*100:.1f}%',  # convert to percentage
                (p.get_x() + p.get_width()/2, height),
                ha='center', va='bottom', fontsize=10, color='black')

plt.title('Survival Rate by Sex (%)',fontsize=14, pad=10, fontweight='bold')
plt.ylabel('Survival Rate')

plt.show()
```

**Evaluation**: Who survived more often, men or women?

- It can be clearly seen that women survived more as compared to men, which mean during the Titanic diaster, women were priortized for lifeboats under the “women and children first” practice.Therefore, being a women had a greater impact on survival rate.


**Key Insight**: Being female was one of the strongest predictors of survival.
                    

```python
plt.figure(figsize=(8,5))
ax=sns.barplot(x='Pclass', y='Survived', data=df, hue='Pclass', palette={1:'#4C72B0', 2:'#8172B3', 3:'#C44E52'},errorbar=None)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height*100:.1f}%',
                (p.get_x() + p.get_width()/2, height),
                ha='center', va='bottom', fontsize=10, color='black')

plt.title('Survival Rate by Passenger class',fontsize=14, pad=10, fontweight='bold')
plt.show()
```

**Evaluation**: How survival vary across passenger class?

- Higher class passengers, which is 1st class, had more chances of survival class which is obvious as they paid more as compared to lower(2nd and 3rd) class passengers. 
- First class passengers had better/luxury cabins and boats were accessible to them.


**Key Insight**: Socioeconomic status (as represented by passenger class) was a strong predictor of survival.

```python
fig, axes=plt.subplots(1,2, figsize=(12,5))

#Survival rate with respect to Sex
sns.barplot(x='Sex', y='Survived',hue='Sex', data=df, ax=axes[0], palette={'male':'#4C72B0', 'female':'#55A868'}, errorbar= None)
axes[0].set_title("Survived by Sex",fontsize=14, pad=10, fontweight='bold')
axes[0].set_ylabel("Survival Rate")

for p in axes[0].patches:
    axes[0].annotate(f"{p.get_height()*100:.1f}%",
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='bottom', fontsize=10, fontweight='bold')


sns.barplot(x='Pclass', y='Survived' , hue="Pclass", data=df, ax=axes[1], palette={1:'#4C72B0', 2:'#8172B3', 3:'#C44E52'}, errorbar=None)
axes[1].legend_.remove()

axes[1].set_title("Survived by Passenger Class",fontsize=14, pad=10, fontweight='bold')
axes[1].set_ylabel("Survival Rate")
for p in axes[1].patches:
     axes[1].annotate(f"{p.get_height()*100:.1f}%",
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```

I have also made a side by side comparison of boths plots.

- Looking at both plots, it can be seen that females survived more despite the class differences women were prioritised.
- Gender had huge gap as compared to the passenger class.


**Key insight**: 
- Passenger class influenced survival but less dramatically than gender.
- Gender was the strongest single predictor of survival, outweighing the impact of class.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Survival by Fare Distribution' ,fontsize=14, pad=10, fontweight='bold')


plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()
```

**Evaluation**:Are higher fares linked to survival chances?

- First-class passengers who paid high fares tended to have better survival rate. 
- The outlier represents premium-paying passengers and even after accounting for extremes, median fare for survivors was considerably higher 
Key Insight: Fare is moderately linked to survival—wealthier passengers generally had better access to lifeboats.


```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Age',hue="Survived",palette={1:'#4C72B0',0:'#C44E52'}, bins=30, kde=True, multiple='stack')
plt.title("Age Distribution of Titanic Passengers")
plt.show()
```

**Evaluation**: What is the age distribution of survivors vs. non-survivors?

- The age distribution of both survivors and non survivors shows roughly a bell shape, peaking in the 20-30 age range.
- While survival does not vary dramatically but there is a slight suggestions that young people(likec children) might have had better survival rates.
Key Insight: Survival was influenced most by sex and class, while age played a smaller role.
